```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv(r'Age.csv')
```

```python
dataset.head(5)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='Age',y='Premium',data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x=dataset[['Age']]
y=dataset['Premium']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearRegression()
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.intercept_
```

```python
lr.coef_
```
