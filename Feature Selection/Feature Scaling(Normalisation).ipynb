```python
# Min-Max Scaling
# Normalisation Technique
# Xnew=Xi-min(X) / max(x)-min(x)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r'Loan.csv')
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.displot(dataset['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_minmax']=ms.transform(dataset[['CoapplicantIncome']])
```

```python
dataset.head(5)
```

```python
plt.figure(figsize=(7,5))
plt.subplot(1,2,1)
plt.title('Before')
sns.displot(dataset=['CoapplicantIncome'])
plt.subplot(1,2,2)
plt.title('After')
sns.displot(dataset=['CoapplicantIncome_minmax'])
plt.show()
```
