```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r'Loan.csv')
```

```python
dataset.head(6)
```

```python
# Feature Scaling Technique
#1. Standarisation
    # Xnew=Xi-Xmean / SD
```

```python
dataset.isnull().sum()
```

```python
dataset['ApplicantIncome'].fillna(dataset['ApplicantIncome'].mean(),inplace=True)
```

```python
sns.displot(dataset['ApplicantIncome'])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome_ss']=pd.DataFrame(ss.transform(dataset[['ApplicantIncome']]),columns=['x'])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.displot(dataset['ApplicantIncome_ss'])
plt.title('Before')

plt.subplot(1,2,2)
sns.displot(dataset['ApplicantIncome'])
plt.title('After')
plt.show()
```
