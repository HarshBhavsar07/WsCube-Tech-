```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'Size': ['s', 'm', 'l', 'xl', 's', 'm', 'l', 's', 's', 'l', 'xl', 'm']})
df.head(3)

```

```python
ord_data=[['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder() # creates an empty OrdinalEncoder object from sklearn.
```

```python
oe
```

```python
# Initialize encoder with this order
oe = OrdinalEncoder(categories=ord_data)

# Fit encoder on your column
oe.fit(df[['Size']])
```

```python
oe.transform() #Transforming your original 'Size' column into numeric codes
```
